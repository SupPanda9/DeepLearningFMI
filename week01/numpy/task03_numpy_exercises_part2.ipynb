#### Task 4: Python vs NumPy

Use pure Python and `numpy` to create a Python list/`numpy`:

- array of $10$ zeros;
- array of $10$ ones;
- array of $10$ fives;
- array of $10$ zeroes, the fifth value of which is $1$;
- array of all the integers from $10$ to $50$;
- array of all the even integers from $10$ to $50$;
- array that is the reversal of the list `[1, 2, 3]` (save these values in an object and use both methods to reverse that object);
- a 3x3 matrix with values ranging from $0$ to $8$;
- a 3x3 identity matrix;
- a 5x5 matrix with values 1,2,3,4 just below the diagonal.

````python
# Your code here
print("Array of 10 zeros:")
p_zeros = [0 for _ in range(10)]
np_zeros = np.zeros(10)
print(f"Python: {p_zeros}")
print(f"Numpy: {np_zeros}\n")

print("Array of 10 ones:")
p_ones = [1 for _ in range(10)]
np_ones = np.ones(10)
print(f"Python: {p_ones}")
print(f"Numpy: {np_ones}\n")

print("Array of 10 fives:")
p_fives = [5 for _ in range(10)]
np_fives = np.full(10, 5)
print(f"Python: {p_fives}")
print(f"Numpy: {np_fives}\n")

print("Array of 10 zeroes, the fifth value of which is 1")
p_five_zeros = [0] * 10
p_five_zeros[4] = 1
np_five_zeros = np.zeros(10)
np_five_zeros[4] = 1
print(f"Python: {p_five_zeros}")
print(f"Numpy: {np_five_zeros}\n")

print("Array with all the integers from 10 to 50")
p_integers = [i for i in range(10, 51)]
np_integers = np.arange(10, 51)
print(f"Python: {p_integers}")
print(f"Numpy: {np_integers}\n")

print("Array with all the even integers from 10 to 50")
p_integers = [i for i in range(10, 51, 2)]
np_integers = np.arange(10, 51, 2)
print(f"Python: {p_integers}")
print(f"Numpy: {np_integers}\n")

print("Array that is the reversal of the list `[1, 2, 3]`")
p_l = [1, 2, 3]
np_l = np.array(p_l)
print(f"Python: {p_l[::-1]}")
print(f"Numpy: {np_l[::-1]}\n")

print("A 3x3 matrix with values ranging from 0 to 8")
p_matrix = [[i + j*3 for i in range(3)] for j in range(3)]
np_matrix = np.arange(9).reshape(3, 3)
print(f"Python: {p_matrix}")
print(f"Numpy: {np_matrix}\n")

print("A 3x3 identity matrix")
p_identity = [[1 if i == j else 0 for i in range(3)] for j in range(3)]
np_identity = np.identity(3)
print(f"Python: {p_identity}")
print(f"Numpy: {np_identity}\n")

print("A 5x5 matrix with values 1,2,3,4 just below the diagonal")
p_below_diag = [[i if i == j - 1 else 0 for i in range(5)] for j in range(5)]
np_below_diag = np.diag([1,2,3,4], k=-1)
print(f"Python: {p_below_diag}")
print(f"Numpy: {np_below_diag}\n")
````

#### Task 10: Reproducibility

When training models, it's often important to generate the same random numbers every time the application is run, even if the machine changes. This ability helps us remove the possibility of creating the situation: *Well, it works on my machine...*.

Our code should work on all machines and produce the same results across different runs.

Go through the [documentation](https://numpy.org/doc/stable/reference/module_structure.html) and answer the below question.

How can we ensure that we always get the same random numbers (even if we switch machines)? Give two examples.

> **Your answer**:

First way is by using a random global seed applicable to all random generation.

It looks something like this np.random.seed(number)

There is a newer approach - we first set a random number generator

rng = np.random.default_rng(number)

and then using it like this

rng.random(...)

This approach is much better because we are not dealing with shared global state.
